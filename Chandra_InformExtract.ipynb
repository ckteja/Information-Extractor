```python
name_ptrn = re.compile(r"^([a-zA-Z]{3,}\s[a-zA-Z]{1,}'?-?[a-zA-Z]{2,}\s?([a-zA-Z]{1,})?)")
mail_ptrn = re.compile(r'[a-zA-Z0-9-_%+.]+@[a-zA-Z-\.]*\.(com)')
mobile_ptrn = re.compile(r'[\+\d]?(\d{2,3}[-\.\s]??\d{2,3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})')
```

```python
for i in range(len(text)):
  #textstr=str(text[i])
  print('--------------------------------------------')
  print('Resume',i+1,': ')
  print('\nName(s): ')
  """textblob=TextBlob(text[i])
  for j in range(len(textblob.words)):
    if(textblob.words[j]=="Name"):
      print(textblob.words[j+1])"""
  
  name = name_ptrn.finditer(text[i])
  for x in name:
      print(x.group(0))
  
  mail=mail_ptrn.finditer(text[i])
  print('\nEmail(s): ')
  for x in mail:
    print(x.group(0))

  mobile_num = mobile_ptrn.finditer(text[i])
  print('\nMobile Number(s): ')
  
  for x in mobile_num:
    print(x.group(0))
```
